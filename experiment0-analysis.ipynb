```python
import pandas as pd
%matplotlib inline
from ast import literal_eval
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = [16, 6]
```

```python
df = pd.read_csv('log-test.txt')
```

```python
df
```

```python
bigList = []
for item in df['Locations in A']: 
    l = literal_eval(item)
    bigList += l
```

```python
l = bigList
```

```python
# Separate them into columns to make things easier.
a = [i[0] for i in l]
b = [i[1] for i in l]
```

```python
df = pd.DataFrame([a, b], index=['start', 'end']).T
```

```python
x = np.linspace(0, 1, 10) # number of zeroes = number of decimal places of precision
```

```python
width=x[1]-x[0]
width
```

```python
def getCounts(val, width, intervals):
    result = intervals.loc[(intervals['end']>val) & \
                           (intervals['start']<val+width)]
    return len(result)
```

```python
y = [getCounts(val, width, df) for val in x]
```

```python
pd.Series(y).plot()
```
