```python
import pandas as pd
%matplotlib inline
from ast import literal_eval
import numpy as np
import re
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = [16, 6]
```

```python
df = pd.read_csv('log-test.txt')
```

```python
def getDate(filename): 
    """
    Generate dates from filenames. 
    """
    m = re.search('_(\d{4})_', filename)
    if m is not None: 
        return int(m.group(1))
    else:
        return None

df['Date'] = df['Text B'].apply(getDate)
```

```python
x = np.linspace(0, 1, 100) # number of zeroes = number of decimal places of precision
```

```python
width=x[1]-x[0]
width
```

```python
def getCounts(val, width, intervals):
    result = intervals.loc[(intervals['end']>val) & \
                           (intervals['start']<val+width)]
    return len(result)
```

```python
y = [getCounts(val, width, df) for val in x]
```
